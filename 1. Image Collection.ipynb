```python
labels = ['window']
number_imgs = 5
```

# 3. Setup Folders

```python
IMAGES_PATH = os.path.join('Tensorflow', 'workspace', 'images', 'collectedimages')
```

```python
if not os.path.exists(IMAGES_PATH):
    if os.name == 'posix':
        !mkdir -p {IMAGES_PATH}
    if os.name == 'nt':
         !mkdir {IMAGES_PATH}
for label in labels:
    path = os.path.join(IMAGES_PATH, label)
    if not os.path.exists(path):
        !mkdir {path}
```

# 4. Process Images

```python
# Loop through different labels
for label in labels:
    # Get a list of all files in the folder
    folder_path = os.path.join(IMAGES_PATH, label)
    
    # Check if the folder exists
    if os.path.exists(folder_path):
        files = os.listdir(folder_path)

        # Loop through the files and perform actions based on conditions
        for file_name in files:
            # Construct the full path for the file
            file_path = os.path.join(folder_path, file_name)

            # Check file size and format
            if os.path.getsize(file_path) < 10 * 1024 or not file_name.lower().endswith((".jpg", ".jpeg")):
                # Remove the file if it's smaller than 10KB or not in JPG/JPEG format
                os.remove(file_path)
                print(f"Removed {file_name}")
            else:
                # Change file extension to ".jpg"
                new_name = f"{label}_{str(uuid.uuid4())}.jpg"

                # Construct the full paths for the old and new names
                old_path = file_path
                new_path = os.path.join(folder_path, new_name)

                # Rename the file
                os.rename(old_path, new_path)
                print(f"Renamed {file_name} to {new_name}")

    else:
        print(f"Folder {label} not found.")

print("Files have been processed.")
```
